```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

### 1b.
Membaca data dari file yang sudah dibuat pada soal 1a dan simpan ke dalam variabel bernama 
data1

```python
df = pd.read_csv("data_ets.csv", sep=";")
```

### 1c. 
Membuat variabel baru bertipe kategori dari data di kolom prestige dengan ketentuan sebagai berikut: (i) apabila nila prestige > 80 maka label bernilai 1, (ii) selain itu maka label bernilai 0. Tambahkan variabel baru tersebut menjadi kolom paling kanan dengan nama kolom prestige_label. Simpan ke dalam variabel bernama data2.

```python
var_baru = [1 if df["Prestige"][i]>80 else 0 for i in range(df.shape[0])]
```

```python
cat = pd.DataFrame(var_baru, columns=["prestige_label"])
```

```python
df_baru = pd.concat([df, cat], axis=1)
```

```python
df_baru
```

```python
df_baru.info()
```

### 1d.
Hitunglah nilai rata-rata dari masing-masing variabel Income, Education, dan Prestige untuk 
setiap kategori (1 atau 0) pada soal 1c


```python
aggregat = df_baru.groupby("prestige_label").agg({"Income":"mean", "Education":"mean", "Prestige":"mean"})
aggregat = aggregat.reset_index()
aggregat
```

### 1e.
Buatlah scatter plot antara pasangan variabel yang mungkin dari variabel Income, Education, dan Prestige dengan membedakan warna plot untuk kategori (1 dan 0, pada soal 1c) dan menambahkan legend sesuai warna kategori

```python
cols = ["Income", "Education", "Prestige"]
```

```python
for col1 in cols:
    for col2 in cols:
        if col1!=col2:
            plt.figure()
            sns.scatterplot(x=df_baru[col1], y=df_baru[col2], hue=df_baru["prestige_label"])
            plt.show()
```

### 2a, 2b, dan 2c
Hapus baris satu, kemudian hitung korelasi antara Income dan Education, Ulangi langkah 2a untuk setiap baris sampai baris terakhir. 

```python
df2 = df[["Income", "Education"]]
```

```python
df2.shape[0]
data_corr = []
for i in range(0, df2.shape[0]):
    data_corr.append(df2.drop(i, axis=0).corr()["Education"]["Income"])
data_corr = pd.DataFrame(data_corr, columns=["corr"])
```

```python
data_corr
```

### 2d
Simpan datakorelasi ke dalam file (dapat bertipe txt, csv, xls, atau xlsx) 

```python
data_corr.to_csv("data_corr.csv", index=False)
```

### 2e.
Buatlah function untuk merangkum pekerjaan di soal 2a – 2d dengan input argument adalah data 
dan output nilai korelasi. Tambahkan satu baris perintah untuk memanggil function yang anda buat.

```python
def korelasi_aneh(data):
    data_corr = []
    for i in range(0, data.shape[0]):
        data_corr.append(data.drop(i, axis=0).corr()[data.columns[1]][data.columns[0]])
    data_corr = pd.DataFrame(data_corr)
    data_corr.to_csv("data_corr.csv", index=False)
    
```
